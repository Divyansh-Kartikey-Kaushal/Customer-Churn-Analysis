# Customer Churn

Importing essential libraries for EDA and Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing csv file by using pandas 

```python
df = pd.read_csv('Customer Churn.csv')
df.head()
```

## Exploratory Data Analysis

Getting information of the data, its columns with Data type

```python
df.info()
```

Replacing blanks in TotalCharges by 0 and its data type as Float

```python
df['TotalCharges'] = df['TotalCharges'].replace(" ", 0)
df['TotalCharges'] = df['TotalCharges'].astype("float")
```

```python
df.info()
```

Checking shape of the Data (rows, columns)

```python
shape = df.shape
print(shape)
```

To get descriptive statistical analysis of data

```python
df.describe()
```

Checking duplicate Values Present in Data on the basis of Unique Value i.e., customerID

```python
df["customerID"].duplicated().sum()
```

Converting the 0 or 1 value of SeniorCitizen to yes or no to make it easier to understand

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
# checking updated data
df.head(30)
```

Count of Churn Customers

```python
ax = sns.countplot(x = 'Churn', data = df, hue = 'Churn')
ax.bar_label(ax.containers[0])
plt.title("Count of Churn Customers")
plt.show()
```

Percentage of Churn Customers

```python
plt.figure(figsize = (4,5))
gb = df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%0.2f%%")
plt.title("% of Churn Customers")
plt.show()
```

Customer Churn on the basis of gender

```python
plt.figure(figsize = (4,5))
sns.countplot(x = 'gender', data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

Count of Senior Citizen Customers Churn

```python
plt.figure(figsize = (4,5))
ax = sns.countplot(x = 'SeniorCitizen', data = df, hue = 'Churn')
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by SeniorCitizen")
plt.show()
```

Percentage of Senior Customers Churn as Compared to Young Customers Churn

```python
gender_churn_counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()
gender_churn_percent = gender_churn_counts.div(gender_churn_counts.sum(axis=1), axis=0) * 100

# Plot
fig, ax = plt.subplots(figsize=(3, 4))
gender_churn_percent.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])

# Annotate with percentages
for i, bars in enumerate(ax.containers):
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_y() + height/2,
                    f'{height:.1f}%', ha='center', va='center', color='white', fontsize=10, fontweight='bold')

# Labels & Title
plt.title("Churn by SeniorCitizen (Percentage)")
plt.ylabel("Percentage")
plt.xticks(rotation=0)
plt.legend(title="Churn", labels=["No", "Yes"], bbox_to_anchor = (1,1))

plt.show()
```

Customer Churn on the basis of their Contract Type

```python
plt.figure(figsize = (4,5))
ax = sns.countplot(x = 'Contract', data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Count of Customers by Contract")
plt.show()
```

Checking different Services provided by service provider and Churn Rate on that service

```python
df.columns.values
```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))  # 3x3 grid
axes = axes.flatten()  # Flatten for easy iteration

# Generate count plots
for i, col in enumerate(columns):
    sns.countplot(x=df[col], ax=axes[i], hue = df["Churn"])
    axes[i].set_title(f"{col} Distribution")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis='x')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

Customer Churn on the basis on Payment Method
